## Load the data

To start, load the dataset `'titanic.csv'` using pandas.

```python
# Your code here
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
df = pd.read_csv("titanic.csv")
df.head()
```

Use the `.info()` method to quickly preview which features have missing data

```python
# Your code here
df.info()
```

## Observe previous measures of centrality

Let's look at the `'Age'` feature. Calculate the mean, median, and standard deviation of this feature. Then plot a histogram of the distribution.

```python
# Your code here
mean_age = df['Age'].mean()
median_age = df['Age'].median()
std_age = df['Age'].std()

# Print the results
print(f"Mean Age: {mean_age:.2f}")
print(f"Median Age: {median_age:.2f}")
print(f"Standard Deviation of Age: {std_age:.2f}")

# Plot a histogram of the 'Age' column
plt.hist(df['Age'], bins=30, edgecolor='k', alpha=0.7)
plt.title('Histogram of Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show();
```

## Impute missing values using the mean 

Fill the missing `'Age'` values using the average age. (Don't overwrite the original data, as we will be comparing to other methods for dealing with the missing values.) Then recalculate the mean, median, and std and replot the histogram.

```python
# Your code here


# Create a copy of the DataFrame
df_filled = df.copy()

# Calculate the mean of the 'Age' column
mean_age = df['Age'].mean()

# Fill missing values in the copied DataFrame with the mean age
df_filled['Age'].fillna(mean_age, inplace=True)

# Recalculate the mean, median, and standard deviation for the filled DataFrame
mean_age_filled = df_filled['Age'].mean()
median_age_filled = df_filled['Age'].median()
std_age_filled = df_filled['Age'].std()

# Print the recalculated statistics
print(f"Mean Age (after filling): {mean_age_filled:.2f}")
print(f"Median Age (after filling): {median_age_filled:.2f}")
print(f"Standard Deviation of Age (after filling): {std_age_filled:.2f}")

# Plot a histogram of the 'Age' column in the filled DataFrame
plt.hist(df_filled['Age'], bins=30, edgecolor='k', alpha=0.7)
plt.title('Histogram of Age Distribution (After Filling Missing Values)')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

```

### Commentary

Note that the standard deviation dropped, the median was slightly raised and the distribution has a larger mass near the center.

## Impute missing values using the median 

Fill the missing `'Age'` values, this time using the median age. (Again, don't overwrite the original data, as we will be comparing to other methods for dealing with the missing values.) Then recalculate the mean, median, and std and replot the histogram.

```python
# Your code here
df_filled_median = df.copy()

# Calculate the median of the 'Age' column
median_age = df['Age'].median()

# Fill missing values in the copied DataFrame with the median age
df_filled_median['Age'].fillna(median_age, inplace=True)

# Recalculate the mean, median, and standard deviation for the filled DataFrame
mean_age_filled_median = df_filled_median['Age'].mean()
median_age_filled_median = df_filled_median['Age'].median()
std_age_filled_median = df_filled_median['Age'].std()

# Print the recalculated statistics
print(f"Mean Age (after filling with median): {mean_age_filled_median:.2f}")
print(f"Median Age (after filling with median): {median_age_filled_median:.2f}")
print(f"Standard Deviation of Age (after filling with median): {std_age_filled_median:.2f}")

# Plot a histogram of the 'Age' column in the filled DataFrame
plt.hist(df_filled_median['Age'], bins=30, edgecolor='k', alpha=0.7)
plt.title('Histogram of Age Distribution (After Filling Missing Values with Median)')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

### Commentary

Imputing the median has similar effectiveness to imputing the mean. The variance is reduced, while the mean is slightly lowered. You can once again see that there is a larger mass of data near the center of the distribution.

## Dropping rows

Finally, let's observe the impact on the distribution if we were to simply drop all of the rows that are missing an age value. Then, calculate the mean, median and standard deviation of the ages along with a histogram, as before.

```python
# Your code here
df_dropped = df.dropna(subset=['Age'])

# Recalculate the mean, median, and standard deviation for the 'Age' column
mean_age_dropped = df_dropped['Age'].mean()
median_age_dropped = df_dropped['Age'].median()
std_age_dropped = df_dropped['Age'].std()

# Print the recalculated statistics
print(f"Mean Age (after dropping missing values): {mean_age_dropped:.2f}")
print(f"Median Age (after dropping missing values): {median_age_dropped:.2f}")
print(f"Standard Deviation of Age (after dropping missing values): {std_age_dropped:.2f}")

# Plot a histogram of the 'Age' column in the DataFrame with dropped rows
plt.hist(df_dropped['Age'], bins=30, edgecolor='k', alpha=0.7)
plt.title('Histogram of Age Distribution (After Dropping Missing Values)')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

### Commentary

Dropping missing values leaves the distribution and associated measures of centrality unchanged, but at the cost of throwing away data.
